# Machine Learning

- **Definition**: Field of study that gives computers the ability to learn without being explicitly programmed.

## Machine Learning Algorithms

- Supervised Learning
  - used most in real-world applications
  - seen rapid advancements
- Unsupervised Learning
- Recommender Systems
- Reinforcement Learning
- _etc._

> **Note**: Supervised, Unsupervised, and Recommender Systems are the most common types of Machine Learning algorithms.

### Supervised Learning

- **Definition**: The machine is trained using labeled data.

  
  

### Types of Supervised Learning

- **Regression**
  - used when the output is a continuous value.

    

- **Classification**
  - used when the output is a category.

    

    

    

- Summary

  

## Unsupervised Learning

- In Unsupervised Learning, the machine is trained using unlabeled data.
- **data only comes with inputs $x$, but not output labels $y$**. Algorithm has to find structure in the data.
  - **clustering**: grouping similar data points together
  - **dimensionality reduction**: compress data using fewer numbers
  - **anomaly detection**: finding unusual data points
- because we are not trying to supervise the algorithm to give some "right answer" for every input, instead we ask the algotithm to find some structure in the data.

  

- eg. _Google News_, _DNA microarray_
  
  

  

### Examples of Supervised and Unsupervised Learning

- Given email labeled as spam/not spam, learn a spam filter **_(Supervised Learning)_**
- Given a set of news articles found on the web, group them into sets of articles about the same story **_(Unsupervised Learning)_**
- Given a database of customer data, automatically discover market segments and group customers into different market segments **_(Unsupervised Learning)_**
- Given a dataset of patients diagnosed as either having diabetes or not having diabetes, learn to classify new patients as having diabetes or not **_(Supervised Learning)_**

# Linear Regression Model

- **Definition**: Linear regression is a linear approach to modeling the relationship between a scalar response (dependent variable) and one or more explanatory variables (independent variables).
- fitting a straight line to the data
- any supervised learning that predicts a number is a **regression problem**
  - infinitely many possible outputs in regression model
- any supervised learning that predicts a category or discrete categories is a **classification problem**
  - small number of possible outputs in classification model

  

  

### Terminology

- **Training Set:** Data used to train the model
- **$x$:** Input variable / features
- **$y$:** Output variable / target variable
- **$m$:** Number of training examples
- **$(x, y)$:** One training example
- **$(x^{(i)}, y^{(i)})$:** $i^{th}$ training example
  - $i$ is the index / row of the training example

> **Note**: $(x^{(i)}, y^{(i)})$ is a single training example, not a power operation.

  

  

### Cost Function

- **Definition**: A function that measures the performance of a machine learning model for given data.
- for linear regression, we use the **mean squared error** cost function _(wil discuss later)_
- for $f(x) = wx + b$, the cost function is $J(w, b)$
